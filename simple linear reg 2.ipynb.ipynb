```python
!pip install plotly
!pip install scikit-learn
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px

import warnings
warnings.filterwarnings('ignore')
```

```python
df=pd.read_csv("height-weight.csv")
df
```

```python
plt.scatter(x=df.Weight,y=df.Height)
plt.show()
```

```python
x=df[["Weight"]] # is a df
y=df['Height'] # is a series
```

```python
y
```

```python
x
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=.25,random_state=3)
```

```python
x_train.shape,x_test.shape,y_train.shape,y_test.shape
```

```python
from sklearn.preprocessing import StandardScaler
sde=StandardScaler()
# scaling always done on input variablesdatframe like x ...not on output/dependent variable
```

```python
x_train_scaled=sde.fit_transform(x_train)   # proper scaling for train vs test data
x_test_scaled=sde.transform(x_test)     # 
```

```python
x_train_scaled
```

```python
x_test_scaled.size
```

```python
plt.scatter(x_train_scaled,y_train)
```

```python
#model training
```

```python
from sklearn.linear_model import LinearRegression
```

```python
regressor=LinearRegression()
```

```python
regressor.fit(x_train_scaled,y_train)
```

```python
regressor.intercept_
print(" the slope or coeff of model theta1 is",regressor.coef_)
```

```python
regressor.coef_
print(" the slope or coeff of model is theta nod",regressor.intercept_)
```

```python
#fit train the model
#predict the model
#after building the model,you have attributes of the model
#coef
#reg
```

```python
y_train_predicted=regressor.predict(x_train_scaled)
y_train_predicted
```

```python
plt.scatter(x_train_scaled,y_train)
plt.plot(x_train_scaled,y_train_predicted,'r')
```

```python
y_predict_test=regressor.predict(x_test_scaled)
```

```python
plt.scatter(x_test_scaled,y_test)
plt.plot(x_test_scaled,y_predict_test,'r')
plt.show()
```

```python
# perfromace matrixes
```

```python
from sklearn.metrics import mean_absolute_error,mean_squared_error,r2_score
```

```python
mean_absolute_error(y_test,y_predict_test)
```

```python
mean_squared_error(y_test,y_predict_test)
```

```python
r2_score(y_test,y_predict_test)
```
