```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px

import warnings
warnings.filterwarnings('ignore')
```

```python
from sklearn.datasets import load_diabetes
```

```python
diabetes=load_diabetes()
```

```python
diabetes
```

```python
print(diabetes.DESCR)
```

```python
diabetes.data
```

```python
diabetes.target
```

```python
diabetes.feature_names
```

```python
df=pd.DataFrame(diabetes.data,columns=diabetes.feature_names)
```

```python
df
```

```python
df['target'] = diabetes.target #important
```

```python
x=df[['bmi']]
x
```

```python
y = df['target'] #dependent variable
```

```python
x
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=4,test_size=.2)
```

```python
y_train
```

```python
y_test
```

```python
x_test
```

```python
x_train
```

```python
from sklearn.linear_model import LinearRegression
model=LinearRegression()
```

```python
model
```

```python
model.fit(x_train,y_train)
```

```python
model.coef_
```

```python
model.intercept_
```

```python
y_predict_test=model.predict(x_test)
```

```python
plt.scatter(x_train,y_train,color='skyblue',label='actual')
plt.plot(x_test,y_predict_test,color='pink',linewidth=3,label="predicted linear regresion line")
plt.xlabel("bmi")
plt.ylabel("one year target")
plt.legend()
plt.show()
```

```python
plt.scatter(x_test,y_test,color='skyblue',label='actual')
plt.plot(x_test,y_predict_test,color='pink',linewidth=3,label="predicted linear regresion line")
plt.xlabel("bmi")
plt.ylabel("one year target")
plt.legend()
plt.show()
```

```python
# model evaluation
```

```python
from sklearn.metrics import mean_squared_error,r2_score,accuracy_score
mean_squared_error(y_test,y_predict_test)
```

```python
r2_score(y_test,y_predict_test)
```
