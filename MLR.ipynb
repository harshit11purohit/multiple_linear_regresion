```python
!pip install plotly
!pip install scikit-learn
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px

import warnings
warnings.filterwarnings('ignore')
```

```python
from sklearn.datasets import load_diabetes
```

```python
db=load_diabetes()
```

```python
df=pd.DataFrame(db.data,columns=db.feature_names)
df
```

```python
df['target']=db.target
df
```

```python
y=df.target
x=df.drop('target',axis=1)
#x=df.iloc[:,:-1]
#y=df.iloc[:-1]
```

```python
x
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=4,test_size=.2)
```

```python
from sklearn.linear_model import LinearRegression
model=LinearRegression()
```

```python
model.fit(x_train,y_train)
```

```python
y_train_predict=model.predict(x_train) # less need
y_test_predict=model.predict(x_test)
```

```python
#plt.scatter(x_train,y_train)
plt.scatter(y_train,y_train_predict)
```

```python
plt.plot(y_test,y_test_predict)
```

```python
model.intercept_
```

```python
model.coef_
```

```python
plt.scatter(y_test, y_test_predict)
plt.plot(
    [y_test.min(), y_test.max()],
    [y_test.min(), y_test.max()],
    'r'
)
plt.xlabel("Actual y (y_test)")
plt.ylabel("Predicted y (ŷ_test)")
plt.title("Actual vs Predicted (MLR)")
plt.show()
```
